```python
import scipy.stats as sts
import numpy as np
import matplotlib.pyplot as plt 
# импортирую нужные библиотеки
```

```python
lamda=2
Pu = sts.poisson(lamda)
```

```python
sample = Pu.rvs(1000) # генерирую выборку размером 1000
```

```python
plt.plot(x,Pu.pmf(x)) # строю теор. функцию плости распр. Пуассона
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.hist(sample,bins =8,density = True) # Гисторгамма от выборки
```

# Для распределения Пуассоона выполено:
# $$mu = lamda$$ 
# $$sigma= lamda$$

# Большое количество выборок Пуассона оценивается 
# $$N(lamda,srqrt(lamda))$$
(Википедия)

```python
def relation(n): # тк нужно несколько раз выводить графики, создам функцию, зависющую от числа выборок
    samples_avg=[] #массив со средними каждой из выборок
    for i in range(1000):
        samples_avg.append(Pu.rvs(n).mean()) 

    mean = 2 
    sigma = np.sqrt(2)
    norm = sts.norm(2,sigma/np.sqrt(n-1)) # второй параметр - это стандартное отклонение
    plt.hist(samples_avg, bins = 5*int(np.sqrt(n)), density = True,label = 'current')
    #для красоты картинок сделал, чтобы количество бинов зависело от n (формула от балды)
    x = np.linspace(-0,5,100)
    plt.plot(x,norm.pdf(x), label = 'theory') 
    plt.legend()
    plt.ylabel('$f(x)$') #навожу красоту
    plt.xlabel('$x$')
    plt.show()
```

```python
relation(3)
```

```python
relation(5)
```

```python
relation(25)
```
